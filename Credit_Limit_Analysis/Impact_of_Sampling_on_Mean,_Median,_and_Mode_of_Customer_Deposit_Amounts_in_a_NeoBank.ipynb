#Objective:
To observe the variability of mean, median, and mode for different sample sizes using Python in Google Colab.

```python
#Load the Dataset
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00350/default%20of%20credit%20card%20clients.xls"
df = pd.read_excel(url, header=1)
```

```python
#Explore the data set
print(df.head())
```

```python
#Ramdon Sampling & Metrics Calculation
sample_sizes = [100, 500, 1000, 5000]

metrics = {
    "Sample_Size": [],
    "Mean": [],
    "Median": [],
    "Moda": []
}

for size in sample_sizes:
  sample = df.sample(size)
  metrics["Sample_Size"].append(size)
  metrics['Mean'].append(sample["LIMIT_BAL"].mean())
  metrics['Median'].append(sample["LIMIT_BAL"].median())
  mode_val = sample["LIMIT_BAL"].mode().iloc[0]
  metrics['Moda'].append(mode_val)

metrics_df = pd.DataFrame(metrics)
print(metrics_df)
```
